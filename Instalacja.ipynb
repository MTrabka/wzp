# Instalacja Pythona

W celu zainstalowania Pythona proszę wejść na stronę https://www.python.org/downloads/

Aby w przyszłości uniknąć problemów z instalacją dodatkowych modułów, proszę zaznaczyć pole jak na rysunku:

# Instalacja Jupyter Notebook

W konsoli proszę wpisać:

`pip install notebook`

Aby uruchomić notebook w konsoli należy wpisać:

`jupyter notebook`

**Uwaga: Podczas pracy w notebooku nie można zamknąć konsoli**
